price prediction using linear regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer

df = pd.read_csv("bersihCarPrice.csv")

X = df[["carwidth", "carlength", "stroke", "highwaympg"]]
y = df["price"]
print (X.dtypes)
imputer = SimpleImputer(strategy='mean')
X = imputer.fit_transform(X)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.4, random_state=22)

model = LinearRegression()
model.fit(X_train, y_train)

model_pred = model.predict(X_test) 

width = float(input("Masukin width dari car: "))
length = float(input("Masukan length nya : "))
stroke = float(input("masukin stroke : "))
highway = float(input("masukin highwaynya : "))

new_data = pd.DataFrame({
    'carwidth':[width],
    'carlength' : [length],
    'stroke' : [stroke],
    'highwaympg' : [highway]
})

predicted_price = model.predict(new_data)
print(f"harganya adalah {predicted_price[0]:.2f}")


# evaluasi
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np
mae = mean_absolute_error(y_test,model_pred)
print(f"MAE = {mae:.2f}")

mse = mean_squared_error(y_test, model_pred)
print(f"MSE = {mse:2f}")

rmse = np.sqrt(mse)
print(f"RMSE : {rmse:.2f}")
```
